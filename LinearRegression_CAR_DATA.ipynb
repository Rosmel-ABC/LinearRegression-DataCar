```python
#Importamos las librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression # entrenar el modelo
from sklearn.model_selection import train_test_split  # dividir los datos
from sklearn import metrics  #metricas de evaluacion
```

```python
data_car = pd.read_csv("/content/car data.csv")
data_car
```

```python
data_car.shape
```

```python
data_car.info()
```

```python
#Veamos si hay valores nulos
data_car.isnull().sum()
```

```python
# Separamos las entradas X y el target y
X = data_car.drop(["Car_Name", "Selling_Price"] , axis=1)
Y = data_car["Selling_Price"]

X
```

##Reemplazamos los datos categÃ³ricos por valores numÃ©ricos

```python
print(X["Fuel_Type"].value_counts())

print(X["Seller_Type"].value_counts())

print(X["Transmission"].value_counts())
```

```python
#Reemplazo de valores

X_input = X.replace(    { "Fuel_Type" : {"Petrol": 0 , "Diesel" : 1, "CNG" : 2} ,
                "Seller_Type" : {"Dealer" : 0 , "Individual": 1  },
                "Transmission" :{"Manual": 0 , "Automatic": 1}
               } )
```

```python
X_input
```

## Dividir el SET  de datos en TRAIN  y TEST

```python
X_train , X_test , y_train , y_test = train_test_split(X_input, Y , test_size = 0.2)
```

## Entrenamos el modelo 

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train , y_train)
```

```python
# prediccion con los datos de entrenamiento
y_pred = regressor.predict(X_test)
```

## Creamos un tabla con los resutados reales y predichos

```python
tabla = pd.DataFrame({"Valores Reales" : y_test , "Valores Predichos": y_pred})
tabla
```

```python
from sklearn  import metrics
error = metrics.mean_squared_error(y_test, y_pred)
print("El error cuadrÃ¡tico medio" , np.sqrt(error))
```
